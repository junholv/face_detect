## Charger les classificateurs en cascade pré-entrainés

```python
#On charge les classificateur
#upload classificator 

#visage 
#faces
face_cascade = cv.CascadeClassifier('haarcascade_frontalface_default.xml')

#pour les yeux :
#for the eyes : 
eyes_cascade = cv.CascadeClassifier('haarcascade_eye.xml')
```

## Charger les images

```python
# on peut charger 1 ou plusieur images
# you can add one or more image

img = cv.imread('dad_mom.jpg')
```

```python
#convertir cette image en niveau de gris
# convert you photo to grey for the classificator 
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
```

## Exécution de la détection de visage

```python
# detectMultiScale est une fonction
#detectMultiScale(image, scale factor, number of neighbors)
# on l'applique sur l'image en niveau de gris
# on va donner une valeur pour le premier paramettre
#une pour le seconde
# 1.1 => 10%
# 3 => paramettre du nombre de voisin

faces = face_cascade.detectMultiScale(gray,1.05, 8)

#pour les yeux
eyes = eyes_cascade.detectMultiScale(gray,1.05, 8)
```

## Affichage des visages

```python
#voir le nombre de visage au nombre de vecteur ici 2
# you can see the number of vector correspond to faces (2)
for face in faces:
    print(face)
```

```python
#la même chose pour les yeux
#the same things like faces
for eye in eyes:
    print(eye)
```

```python
# (x,y) = cordonné du point en haut a haut a gauche
# (x+w, y+h) = coordonné du point en bas a droite
#(0,255,0) = BGR ; ici Green = 255
# 2 => epaisseur de la ligne du rectangle
i=0
# affichage des visages
# show faces
for face in faces:
    x,y,w,h = face
    
    #dessiner le rectangle sur l'image
    #draw rectangle around faces
    cv.rectangle(img,(x,y), (x+w, y+h), (0,255,0),2)

# afficher des yeux
# highlighting eyes
for (ex, ey, ew, eh) in eyes:
#     dessiner le rectangle autour des yeux sur l'iamge princiaple
#     draw rectangle around eyes
    cv.rectangle(img, (ex,ey), (ex + ew, ey + eh), (255,0,0),2)
```

```python
# affiche l'image principal:
#highlighting of the principal photo
# cv.imshow('image princiaple', img)

#sauvegarder l'image 
#save picture
cv.imwrite( "dad_mom_faces.jpg", img )


#cv.destroyAllWindows()
```
